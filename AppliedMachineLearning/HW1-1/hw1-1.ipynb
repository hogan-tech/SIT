## Question 4
For the previous generated data, consider each row of the array as an individual data point, with the two columns representing its distinct features. Identify the data points with the smallest and largest sum of features. Mark them in red and blue in the figure respectively.

Hint: You can compute the sum of each row and store it in an array named "sums". Then find the minimum or maximum index of this array. 

```python
# Generate a random array with 100 rows and 2 columns
data = np.random.rand(100, 2)

# Compute the sum of each row and store it in an array named "sums"
sums = np.sum(data, axis=1)

# Find the indices of the minimum and maximum sums
smallestSum = np.argmin(sums)
largestSum = np.argmax(sums)

# Plotting the data
# Set the figure size
plt.figure(figsize=(8, 6))

plt.scatter(data[:, 0], data[:, 1], color='black', label='Data Points')

# Mark the point with the smallest sum in red
plt.scatter(data[smallestSum, 0], data[smallestSum, 1],
            color='red', label='Smallest Sum')

# Mark the point with the largest sum in blue
plt.scatter(data[largestSum, 0], data[largestSum, 1],
            color='blue', label='Largest Sum')

plt.title('Data Points with Extremes Marked', fontsize=16)
plt.xlabel('Feature 1', fontsize=12)
plt.ylabel('Feature 2', fontsize=12)

# Add legend
# Check from this web site https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
# Put the legent entry in one line https://stackoverflow.com/questions/43644758/how-to-put-all-legend-entries-on-one-line
# Customize the frame https://jakevdp.github.io/PythonDataScienceHandbook/04.06-customizing-legends.html
plt.legend(loc=8, borderaxespad=-6, ncol=3, fancybox=True,
           framealpha=1, shadow=True, borderpad=0.5)

# Show plot
plt.show()
```

## Question 5
Add two lines representing the mean of the Feature 1 and the 25% quantile of the Feature 2 seperately. Add the legend of the two lines.

```python

# Compute the mean
meanFeature1 = np.mean(data[:, 0])

# Compute the 25% quantile
quantile25Feature2 = np.percentile(data[:, 1], 25)

# Add vertical line
plt.axvline(x=meanFeature1, color='green', linestyle='-', label='Mean of Col 1')

# Add horizontal line
plt.axhline(y=quantile25Feature2, color='orange', linestyle='-', label='25% Quantile of Col 2')

# Scatter plot for all points
plt.scatter(data[:, 0], data[:, 1], color='black', label='Data Points')

# Add title and labels
plt.title('Scatter Plot with Mean and Quantile Lines', fontsize=16)
plt.xlabel('Feature 1', fontsize=12)
plt.ylabel('Feature 2', fontsize=12)

# Add legend
# Check from this web site https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
# Put the legent entry in one line https://stackoverflow.com/questions/43644758/how-to-put-all-legend-entries-on-one-line
# Customize the frame https://jakevdp.github.io/PythonDataScienceHandbook/04.06-customizing-legends.html
plt.legend(loc=8, borderaxespad=-6, ncol=3, fancybox=True,
           framealpha=1, shadow=True, borderpad=0.5)

# Show plot
plt.show()
```

## Question 6
Display the figures in Quesiton 3-5 in one figure. Please follow the given output to adjust the arangement . So that you can observe and compare the figures more convenienty. Don't forget parts of the original figures change.

Note that the first subplot is given. You just need to plot the other three figures. 

```python
# Generate a random array with 100 rows and 2 columns
data = np.random.rand(100, 2)

# Compute the sum of each row and store it in an array named "sums"
sums = np.sum(data, axis=1)

# Find the indices of the minimum and maximum sums
smallestSum = np.argmin(sums)
largestSum = np.argmax(sums)

# Compute the mean of Feature 1 (first column)
meanFeature1 = np.mean(data[:, 0])

# Compute the 25% quantile of Feature 2 (second column)
quantile25Feature2 = np.percentile(data[:, 1], 25)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# First Subplot (Original scatter plot from Question 3)
axes[0, 0].scatter(data[:, 0], data[:, 1],
                   color='black', label='Random Points')
axes[0, 0].set_title(
    'Original Data Points', fontsize=14)
axes[0, 0].set_xlabel('Feature 1', fontsize=12)
axes[0, 0].set_ylabel('Feature 2', fontsize=12)

# Second Subplot (Mark min and max points from Question 4)
axes[0, 1].scatter(data[:, 0], data[:, 1], color='black', label='Data Points')
axes[0, 1].scatter(data[smallestSum, 0], data[smallestSum, 1],
                   color='red', label='Smallest Sum')
axes[0, 1].scatter(data[largestSum, 0], data[largestSum, 1],
                   color='blue', label='Largest Sum')
axes[0, 1].set_title('Data Points with Extremes Marked', fontsize=16)
axes[0, 1].set_xlabel('Feature 1', fontsize=12)
axes[0, 1].set_ylabel('Feature 2', fontsize=12)
axes[0, 1].legend(loc="lower right")

# Third Subplot (Mean lines from Question 5)
axes[1, 0].scatter(data[:, 0], data[:, 1], color='black')
axes[1, 0].axvline(x=meanFeature1, color='green',
                   linestyle='-', label='Mean of Col 1')
axes[1, 0].set_title('Scatter Plot with Mean and Quantile Lines', fontsize=16)
axes[1, 0].set_xlabel('Feature 1', fontsize=12)
axes[1, 0].set_ylabel('Feature 2', fontsize=12)
axes[1, 0].legend(loc="lower right")

# Fourth Subplot (25% Quantile lines from Question 5)
axes[1, 1].scatter(data[:, 0], data[:, 1], color='black')
axes[1, 1].axhline(y=quantile25Feature2, color='orange',
                   linestyle='-', label='25% Quantile of Col 2')
axes[1, 1].set_title('Scatter Plot with Mean and Quantile Lines', fontsize=16)
axes[1, 1].set_xlabel('Feature 1', fontsize=12)
axes[1, 1].set_ylabel('Feature 2', fontsize=12)
axes[1, 1].legend(loc="lower right")

# Adjust the layout to prevent overlap of titles and labels
plt.tight_layout()

# Show the final plot
plt.show()
```
